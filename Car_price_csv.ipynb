## 2-How do sales vary over time (e.g., monthly, quarterly, yearly)?
Visualization: Line plot or bar chart over time

```python
df.drop(columns=["seller", "trim","condition","odometer","state","vin"], inplace=True)
```

```python
# if df["saledate"].dtype != "str":
#   df["saledate"] = df["saledate"].astype("str")
# df['date_column'] = pd.to_datetime(df['saledate'])
# df.drop(columns="saledate",inplace=True)
# df.head()
```

```python
df["transmission"].value_counts()
```

```python
df["saledate"]= df["saledate"].astype("str")
df.info()
```

```python
df["saledate"].str.split(" ",expand=True)[0]
```

```python
df["saledate"] = pd.to_datetime(df["saledate"],utc=True)
df.info()
```

```python
df["Selling_Year"] = df["saledate"].dt.year
df["Selling_Month"] = df["saledate"].dt.month
# df.drop(columns="Year",inplace=True)
```

```python
df["Selling_Year"].dropna(inplace=True)
df["Selling_Year"].isnull().sum()
```

```python
df['Selling_Year'] = df['Selling_Year'].replace([float('inf'), float('-inf')], pd.NA).fillna(0).astype(int)
df['Selling_Month'] = df['Selling_Month'].replace([float('inf'), float('-inf')], pd.NA).fillna(0).astype(int)
```

```python
df.head()
```

```python
Selling_mean_Month=df.groupby("Selling_Month")["sellingprice"].mean()
```

```python
plt.plot(Selling_mean_Month, marker="*",color='orange')
plt.title('Sales Over Month')
plt.xlabel('Month')
plt.ylabel('Sales');
```

## 3-What is the relationship between selling price and mileage?
Visualization: Scatter plot

```python
correlation = df["sellingprice"].corr(df["mmr"])
plt.figure(figsize=(15,6))
plt.scatter(df["sellingprice"],df["mmr"],alpha=0.5)
plt.title('Relationship between Selling Price and Mileage')
plt.xlabel('Mileage')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show();
```

## 4-Which car make has the highest average selling price?
Visualization: Bar chart or box plot

```python
average_prices=df.groupby("Selling_Year")["sellingprice"].mean().sort_values()
```

```python
plt.figure(figsize=(10, 6))
average_prices.plot(kind='bar', color='skyblue')
plt.title('Average Selling Price by Car Make')
plt.xlabel('Car Make')
plt.ylabel('Average Selling Price')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
